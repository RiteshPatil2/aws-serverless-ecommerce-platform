```python
import pandas as pd

# Load the dataset with exact column names
df = pd.read_csv('2016.csv')

# Check the first few rows of the dataset
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Check for duplicate rows
print(df.duplicated().sum())

# Summary statistics
print(df.describe())

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of DEP_DELAY
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='DEP_DELAY', bins=20, kde=True)
plt.title('Distribution of Departure Delay')
plt.xlabel('Departure Delay')
plt.ylabel('Frequency')
plt.show()

# Scatter plot of DISTANCE vs ARR_DELAY
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='DISTANCE', y='ARR_DELAY')
plt.title('Relationship between Distance and Arrival Delay')
plt.xlabel('Distance')
plt.ylabel('Arrival Delay')
plt.show()


```

```python
#EDA

import pandas as pd
import matplotlib.pyplot as plt

# Load the combined dataset
combined_df = pd.concat([pd.read_csv('2016.csv'), pd.read_csv('2017.csv'), pd.read_csv('2018.csv')])

# 1. Trend Analysis of Flight Delays Over Years
# Calculate average departure delay and arrival delay for each year
yearly_avg_delay = combined_df.groupby('FL_DATE')[['DEP_DELAY', 'ARR_DELAY']].mean().reset_index()


# Plot trend of average delay over the three years
plt.figure(figsize=(10, 6))
plt.plot(yearly_avg_delay['FL_DATE'], yearly_avg_delay['DEP_DELAY'], label='Average Departure Delay')
plt.plot(yearly_avg_delay['FL_DATE'], yearly_avg_delay['ARR_DELAY'], label='Average Arrival Delay')
plt.xlabel('Year')
plt.ylabel('Average Delay (minutes)')
plt.title('Trend of Flight Delays Over Years')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()





```

```python

# 2. Comparison of Airlines' Performance
# Calculate average departure delay, arrival delay, and cancellation rates for each airline
airline_performance = combined_df.groupby('OP_CARRIER')[['DEP_DELAY', 'ARR_DELAY', 'CANCELLED']].mean().reset_index()

# Plot comparison of airlines' performance
plt.figure(figsize=(12, 6))
plt.bar(airline_performance['OP_CARRIER'], airline_performance['DEP_DELAY'], label='Average Departure Delay')
plt.bar(airline_performance['OP_CARRIER'], airline_performance['ARR_DELAY'], label='Average Arrival Delay')
plt.bar(airline_performance['OP_CARRIER'], airline_performance['CANCELLED']*100, label='Cancellation Rate (%)')
plt.xlabel('Airline')
plt.ylabel('Metrics')
plt.title("Comparison of Airlines' Performance")
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
#2

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Filter the dataframe for flights departing from a specific airport
airport_code = 'JFK'
departures_from_airport = df[df['ORIGIN'] == airport_code]

# Calculate the percentage of on-time departures
total_departures = len(departures_from_airport)
on_time_departures = len(departures_from_airport[departures_from_airport['DEP_DELAY'] <= 0])
on_time_percentage = (on_time_departures / total_departures) * 100
total_departures
on_time_departures

# Print the on-time performance
print(f"On-time performance of flights departing from {airport_code}: {on_time_percentage:.2f}%")

# Visualize the insights with a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=['On-time Departures', 'Delayed Departures'], y=[on_time_departures, total_departures - on_time_departures])
plt.title(f"On-time Performance of Flights Departing from {airport_code}")
plt.xlabel("Departure Status")
plt.ylabel("Number of Flights")
plt.show()
```

```python
###3
# Group the data by airport and count the number of cancellations for each airport
airport_cancellations = df[df['CANCELLED'] == 1].groupby('ORIGIN').size().reset_index(name='Cancellation Count')

# Sort the airports by the number of cancellations in descending order
top_10_airports = airport_cancellations.sort_values(by='Cancellation Count', ascending=False).head(10)

# Plot the top 10 airports with the highest number of cancellations
plt.figure(figsize=(10, 6))
sns.barplot(x='ORIGIN', y='Cancellation Count', data=top_10_airports, palette='viridis')
plt.title('Top 10 Airports with Highest Flight Cancellations')
plt.xlabel('Airport')
plt.ylabel('Number of Cancellations')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select relevant columns for weather conditions and flight delays
weather_delays = df[['WEATHER_DELAY', 'DEP_DELAY', 'ARR_DELAY']]

# Remove rows with missing values in weather delay column
weather_delays = weather_delays.dropna(subset=['WEATHER_DELAY'])

# Plot the relationship between weather delay and departure delay
plt.figure(figsize=(10, 6))
sns.scatterplot(x='WEATHER_DELAY', y='DEP_DELAY', data=weather_delays)
plt.title('Relationship between Weather Delay and Departure Delay')
plt.xlabel('Weather Delay (minutes)')
plt.ylabel('Departure Delay (minutes)')
plt.grid(True)
plt.show()

# Plot the relationship between weather delay and arrival delay
plt.figure(figsize=(10, 6))
sns.scatterplot(x='WEATHER_DELAY', y='ARR_DELAY', data=weather_delays)
plt.title('Relationship between Weather Delay and Arrival Delay')
plt.xlabel('Weather Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True)
plt.show()
```
